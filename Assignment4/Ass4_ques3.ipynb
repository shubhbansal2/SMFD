```python
import random
import math
from matplotlib import pyplot as plt
import numpy as np
```

```python
n=100000   # no. of simulations
d=10.0     # distance between the lines
l=5.0      # length of the stick
crossing=0
ans=[]

for i in range(0,n):
    x= random.uniform(0.0,d/2)
    theta= random.uniform(0,math.pi/2)
    if((l/2)* math.sin(theta))>=x:
        crossing+=1;
    ans.append(((2*n*l)/(d*crossing)) if crossing!=0 else 0);
```

```python
print("Final PI estimation:",ans[n-1]);

print("Absolute error:",(abs(math.pi-ans[n-1])))
print("Absolute percentage  error:",(abs(math.pi-ans[n-1])/math.pi)*100,"%")
```

```python
plt.figure(figsize=(12, 6))
plt.semilogx(np.arange(1, n + 1), ans, label='Estimated ')
plt.axhline(y=np.pi, color='r', linestyle='--', label='True ')
plt.xlabel('Number of needle drops (log scale)')
plt.ylabel('Estimated value of ')
plt.title(f'Buffon\'s Needle: Monte Carlo Estimation of \nFinal estimate: {ans[n-1]:.5f} (True : {np.pi:.5f})')
plt.legend()
plt.grid(True, which="both", ls="--")
plt.show()
```

# Error Sources in Buffon's Needle Experiment

## 1. Random Seed Variation
- Different simulation runs will produce slightly different results due to random sampling

## 2. Sample Size Effects
- Larger number of trials (N) gives better estimates (law of large numbers)
- Convergence rate follows ~1/N scaling

## 3. Angle Generation Granularity
- Limited by floating-point precision in   Uniform[0,/2]
- Affects precision of sin() calculations

## 4. Needle Length to Line Distance Ratio
- Optimal estimation efficiency occurs when   d
- Smaller /d ratios require more trials for same precision d")
